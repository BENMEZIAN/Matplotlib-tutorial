# What's subplot
#### In Matplotlib, a subplot is a way of organizing multiple plots in a single figure. You can create subplots in a grid-like layout. The function used to create subplots is plt.subplot() or plt.subplots(). Here's a quick overview of both:

#1. Using plt.subplot()
The plt.subplot() function takes three arguments:
- ***nrows: Number of rows of subplots.***
- ***ncols: Number of columns of subplots.***
- ***index: Index of the subplot you want to access (starting from 1).***

```python
import matplotlib.pyplot as plt

# Create a figure with 2 rows and 2 columns of subplots with plt.subplot(nrows,ncols,index)
plt.subplot(2, 2, 1)  # Row 1, Column 1
plt.plot([1, 2, 3], [1, 4, 9])
plt.title('Plot 1')

plt.subplot(2, 2, 2)  # Row 1, Column 2
plt.plot([1, 2, 3], [1, 2, 3])
plt.title('Plot 2')

plt.subplot(2, 2, 3)  # Row 2, Column 1
plt.plot([1, 2, 3], [9, 4, 1])
plt.title('Plot 3')

plt.subplot(2, 2, 4)  # Row 2, Column 2
plt.plot([1, 2, 3], [3, 2, 1])
plt.title('Plot 4')

plt.tight_layout()  # Adjust the layout so titles don't overlap
plt.show()
```

#### Subplots with 3 rows and 2 columns

```python
# prompt: Subplots 3 rows and 2 columns

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(16, 9))

# Create a figure with 3 rows and 2 columns of subplots
plt.subplot(3, 2, 1)  # Row 1, Column 1
plt.pie([1, 2, 3], labels=["Banana","Orange","Grape"])
plt.legend(["Banana","Orange","Grape"])
plt.title('Plot 1')

plt.subplot(3, 2, 2)  # Row 1, Column 2
plt.hist([1, 2, 3], [1, 2, 3])
plt.title('Plot 2')

plt.subplot(3, 2, 3)  # Row 2, Column 1
plt.scatter([1, 2, 3], [9, 4, 1])
plt.title('Plot 3')

plt.subplot(3, 2, 4)  # Row 2, Column 2
plt.scatter([1, 2, 3], [3, 2, 1])
plt.title('Plot 4')

plt.subplot(3, 2, 5)  # Row 3, Column 1
plt.pie([1, 2, 3])
plt.title('Plot 5')

plt.subplot(3, 2, 6)  # Row 3, Column 2
plt.scatter([1, 2, 3], [6, 5, 4])
plt.title('Plot 6')

plt.tight_layout()  # Adjust the layout so titles don't overlap
plt.show()
```

# 2. Using plt.subplots()

```python
import matplotlib.pyplot as plt

# Create a figure with 2 rows and 2 columns of subplots
fig, axs = plt.subplots(2, 2)

# Plot in the first subplot (row 0, column 0)
axs[0, 0].plot([1, 2, 3], [1, 4, 9])
axs[0, 0].set_title('Plot 1')

# Plot in the second subplot (row 0, column 1)
axs[0, 1].plot([1, 2, 3], [1, 2, 3])
axs[0, 1].set_title('Plot 2')

# Plot in the third subplot (row 1, column 0)
axs[1, 0].plot([1, 2, 3], [9, 4, 1])
axs[1, 0].set_title('Plot 3')

# Plot in the fourth subplot (row 1, column 1)
axs[1, 1].plot([1, 2, 3], [3, 2, 1])
axs[1, 1].set_title('Plot 4')

# Adjust the layout so titles don't overlap
plt.tight_layout()
plt.show()
```
