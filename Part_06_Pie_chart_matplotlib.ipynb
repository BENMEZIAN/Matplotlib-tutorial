```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

fig = plt.figure(figsize=(10, 7))
plt.subplot(1, 2, 1)
plt.pie(y, labels = mylabels)
plt.subplot(1, 2, 2)
plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

fig = plt.figure(figsize=(10, 7))
plt.subplot(1, 2, 1)
plt.pie(y, labels = mylabels)
plt.subplot(1, 2, 2)
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

fig = plt.figure(figsize=(10, 7))
plt.subplot(1, 2, 1)
plt.pie(y, labels = mylabels, radius = 1.4)
plt.subplot(1, 2, 2)
plt.pie(y, labels = mylabels, explode = myexplode, shadow = True, radius = 1.4)
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["green", "yellow", "red", "brown"]
myexplode = [0.2, 0, 0, 0]
textprops = {"fontsize":15}

fig = plt.figure(figsize=(10, 7))
plt.subplot(1, 2, 1)
plt.pie(y, labels = mylabels, colors=mycolors, autopct="%0.1f%%", textprops=textprops)
plt.subplot(1, 2, 2)
plt.pie(y, labels = mylabels, colors=mycolors, autopct="%0.1f%%", textprops=textprops, explode = myexplode, shadow = True)
plt.legend(loc=2, title = "Four Fruits:")
plt.show()
```
