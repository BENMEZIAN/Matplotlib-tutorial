# What's bar plot
#### A bar plot (or bar chart) in Matplotlib is a great way to visualize categorical data. Each bar represents the frequency (or value) of a specific category. Bar plots can be drawn using plt.bar(), which takes the categories on the x-axis and the corresponding values on the y-axis.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [25, 30, 40, 20]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

# Example

```python
import matplotlib.pyplot as plt

# Data: Fruit categories and their respective counts
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes']
counts = [25, 40, 10, 30]

# Create the bar plot
plt.bar(fruits, counts, color='skyblue')

# Add title and labels
plt.title('Fruit Sales', fontsize=16)
plt.xlabel('Fruit Type')
plt.ylabel('Number of Sales')

# Show the plot
plt.show()
```

#Multiple bar plots at the same figure

```python
import numpy as np
import matplotlib.pyplot as plt

barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1, IT, color ='r', width = barWidth, edgecolor ='grey', label ='IT')
plt.bar(br2, ECE, color ='g', width = barWidth, edgecolor ='grey', label ='ECE')
plt.bar(br3, CSE, color ='b', width = barWidth, edgecolor ='grey', label ='CSE')

plt.xlabel('Branch', fontweight ='bold', fontsize = 15)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IT))], ['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show()
```

# Stacked bar

```python
# prompt: What's the purpose of stacked bar

# Stacked Bar Plot Example

import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Product A', 'Product B', 'Product C']
region1_sales = [25, 30, 40]
region2_sales = [15, 20, 25]

# Create the stacked bar plot
plt.bar(categories, region1_sales, label='Region 1')
plt.bar(categories, region2_sales, bottom=region1_sales, label='Region 2')

# Add title and labels
plt.title('Sales by Region and Product')
plt.xlabel('Products')
plt.ylabel('Sales')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5

boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)

ind = np.arange(N)
width = 0.35

fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, boys, width)
p2 = plt.bar(ind, girls, width,bottom = boys)

plt.ylabel('Contribution')
plt.title('Contribution by the teams (boys and girls in the team)')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data: Fruit categories and their respective counts
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes']
counts = [25, 40, 10, 30]

# Error values (example: ±5 error for each fruit)
error = [5, 5, 3, 4]

# Create the bar plot with error bars
bars = plt.bar(fruits, counts, color='skyblue', yerr=error, capsize=5)

# Add title and labels
plt.title('Fruit Sales', fontsize=16)
plt.xlabel('Fruit Type')
plt.ylabel('Number of Sales')

# Add annotations (number of sales on top of each bar)
for i, count in enumerate(counts):
    print(i,count)
    plt.text(i, count + 1, str(count), ha='center', va='bottom', fontsize=12)

# Add grid lines on the y-axis to improve readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optionally, add a trend line (in this case, for demonstration we can fit a simple line)
# x-axis positions for the bars
x_positions = np.arange(len(fruits))

# Fit a simple line (for demonstration purposes)
coeffs = np.polyfit(x_positions, counts, 1)  # Fit a linear line
trend_line = np.polyval(coeffs, x_positions)

# Plot the trend line over the bars
plt.plot(fruits, trend_line, color='red', linestyle='-', marker='o', label='Trend Line')

# Show the legend for the trend line
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
courses = ['Python', 'R', 'AI', 'ML', 'DL']
class1_students = [30, 10, 20, 25, 13]
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 10]

# Error values (example: ±5 error for each class)
error_class1 = [5, 2, 3, 4, 3]
error_class2 = [4, 1, 2, 3, 2]
error_class3 = [3, 1, 3, 2, 1]

# Create the stacked bar plot
bars1 = plt.bar(courses, class1_students, label='Class 1', color='skyblue', yerr=error_class1, capsize=5)
bars2 = plt.bar(courses, class2_students, bottom=class1_students, label='Class 2', color='lightgreen', yerr=error_class2, capsize=5)
bars3 = plt.bar(courses, class3_students, bottom=[i+j for i,j in zip(class1_students, class2_students)], label='Class 3', color='salmon', yerr=error_class3, capsize=5)

# Add title and labels
plt.title('Number of Students by Each Class', fontsize=16)
plt.xlabel('Courses Offered')
plt.ylabel('Number of Students')

# Add annotations (number of students on top of each segment)
for i, (class1, class2, class3) in enumerate(zip(class1_students, class2_students, class3_students)):
    plt.text(i, class1 + 2, str(class1), ha='center', va='bottom', fontsize=12)
    plt.text(i, class1 + class2 + 2, str(class2), ha='center', va='bottom', fontsize=12)
    plt.text(i, class1 + class2 + class3 + 2, str(class3), ha='center', va='bottom', fontsize=12)

# Add grid lines on the y-axis to improve readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Optionally, add a trend line (in this case, we can fit a simple line for demonstration)
# x-axis positions for the bars
x_positions = np.arange(len(courses))

# Calculate the total number of students in each course (stacked sum)
total_students = [i + j + k for i, j, k in zip(class1_students, class2_students, class3_students)]

# Fit a simple line (for demonstration purposes)
coeffs = np.polyfit(x_positions, total_students, 1)  # Fit a linear line
trend_line = np.polyval(coeffs, x_positions)

# Plot the trend line over the bars
plt.plot(courses, trend_line, color='red', linestyle='-', marker='o', label='Trend Line')

# Show the legend for the trend line
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
courses = ['Python', 'R', 'AI', 'ML', 'DL']
class1_students = [30, 10, 20, 25, 13]
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 10]

# Error values (example: ±5 error for each class)
error_class1 = [5, 2, 3, 4, 3]
error_class2 = [4, 1, 2, 3, 2]
error_class3 = [3, 1, 3, 2, 1]

# Create the horizontal stacked bar plot
bars1 = plt.barh(courses, class1_students, label='Class 1', color='skyblue', xerr=error_class1, capsize=5)
bars2 = plt.barh(courses, class2_students, left=class1_students, label='Class 2', color='lightgreen', xerr=error_class2, capsize=5)
bars3 = plt.barh(courses, class3_students, left=[i+j for i,j in zip(class1_students, class2_students)], label='Class 3', color='salmon', xerr=error_class3, capsize=5)

# Add title and labels
plt.title('Number of Students by Each Class', fontsize=16)
plt.ylabel('Courses Offered')
plt.xlabel('Number of Students')

# Add annotations (number of students on top of each segment)
for i, (class1, class2, class3) in enumerate(zip(class1_students, class2_students, class3_students)):
    plt.text(class1 + 2, i, str(class1), ha='left', va='center', fontsize=12)
    plt.text(class1 + class2 + 2, i, str(class2), ha='left', va='center', fontsize=12)
    plt.text(class1 + class2 + class3 + 2, i, str(class3), ha='left', va='center', fontsize=12)

# Add grid lines on the x-axis to improve readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Optionally, add a trend line (in this case, we can fit a simple line for demonstration)
# y-axis positions for the bars (using course indices)
y_positions = np.arange(len(courses))

# Calculate the total number of students in each course (stacked sum)
total_students = [i + j + k for i, j, k in zip(class1_students, class2_students, class3_students)]

# Fit a simple line (for demonstration purposes)
coeffs = np.polyfit(total_students, y_positions, 1)  # Fit a linear line
trend_line = np.polyval(coeffs, total_students)

# Plot the trend line over the bars
plt.plot(trend_line, y_positions, color='red', linestyle='-', marker='o', label='Trend Line')

# Show the legend for the trend line
plt.legend()

# Show the plot
plt.show()
```
