# What's bar plot
#### A bar plot (or bar chart) in Matplotlib is a great way to visualize categorical data. Each bar represents the frequency (or value) of a specific category. Bar plots can be drawn using plt.bar(), which takes the categories on the x-axis and the corresponding values on the y-axis.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [25, 30, 40, 20]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

# Example

```python
import matplotlib.pyplot as plt

# Data: Fruit categories and their respective counts
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes']
counts = [25, 40, 10, 30]

# Create the bar plot
plt.bar(fruits, counts, color='skyblue')

# Add title and labels
plt.title('Fruit Sales', fontsize=16)
plt.xlabel('Fruit Type')
plt.ylabel('Number of Sales')

# Show the plot
plt.show()
```

# Bar plot with custom colors

```python
import matplotlib.pyplot as plt

# Data
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Watermelon']
counts = [25, 40, 10, 30, 50]

# Create the bar plot with custom colors
colors = ['red', 'yellow', 'orange', 'purple', 'darkgreen']
plt.bar(fruits, counts, color=colors)

# Add title and labels
plt.title('Fruit Sales')
plt.xlabel('Fruit Type')
plt.ylabel('Number of Sales')

# Show the plot
plt.show()
```

# Horizontal bar plot

```python
import matplotlib.pyplot as plt

# Data
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Watermelon']
counts = [25, 40, 10, 30,50]

# Create a horizontal bar plot
plt.barh(fruits, counts, color='lightgreen')

# Add title and labels
plt.title('Fruit Sales (Horizontal)')
plt.xlabel('Number of Sales')
plt.ylabel('Fruit Type')

# Show the plot
plt.show()
```

#Adding grid and Customizing Bar Width

```python
import matplotlib.pyplot as plt

# Data
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes']
counts = [25, 40, 10, 30]

# Create the bar plot with custom width and color
plt.bar(fruits, counts, color='orange', width=0.8)

# Add title, labels, and grid
plt.title('Fruit Sales')
plt.xlabel('Fruit Type')
plt.ylabel('Number of Sales')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

# Adding grid and Customizing Bar Height for horizontal

```python
import matplotlib.pyplot as plt

# Data
fruits = ['Apples', 'Bananas', 'Oranges', 'Grapes', 'Watermelon']
counts = [25, 40, 10, 30,50]

# Create a horizontal bar plot
plt.barh(fruits, counts, color='lightgreen', height=0.3)

# Add title and labels
plt.title('Fruit Sales (Horizontal)')
plt.xlabel('Number of Sales')
plt.ylabel('Fruit Type')

# Show the plot
plt.show()
```

#Multiple bar plots at the same figure

```python
import numpy as np
import matplotlib.pyplot as plt

barWidth = 0.25
fig = plt.subplots(figsize =(12, 8))

IT = [12, 30, 1, 8, 22]
ECE = [28, 6, 16, 5, 10]
CSE = [29, 3, 24, 25, 17]

br1 = np.arange(len(IT))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]

plt.bar(br1, IT, color ='r', width = barWidth, edgecolor ='grey', label ='IT')
plt.bar(br2, ECE, color ='g', width = barWidth, edgecolor ='grey', label ='ECE')
plt.bar(br3, CSE, color ='b', width = barWidth, edgecolor ='grey', label ='CSE')

plt.xlabel('Branch', fontweight ='bold', fontsize = 15)
plt.ylabel('Students passed', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(IT))], ['2015', '2016', '2017', '2018', '2019'])

plt.legend()
plt.show()
```

```python
# prompt: What's the purpose of stacked bar

# Stacked Bar Plot Example

import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['Product A', 'Product B', 'Product C']
region1_sales = [25, 30, 40]
region2_sales = [15, 20, 25]

# Create the stacked bar plot
plt.bar(categories, region1_sales, label='Region 1')
plt.bar(categories, region2_sales, bottom=region1_sales, label='Region 2')

# Add title and labels
plt.title('Sales by Region and Product')
plt.xlabel('Products')
plt.ylabel('Sales')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5

boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)

ind = np.arange(N)
width = 0.35

fig = plt.subplots(figsize =(10, 7))
p1 = plt.bar(ind, boys, width)
p2 = plt.bar(ind, girls, width,bottom = boys)

plt.ylabel('Contribution')
plt.title('Contribution by the teams (boys and girls in the team)')
plt.xticks(ind, ('T1', 'T2', 'T3', 'T4', 'T5'))
plt.yticks(np.arange(0, 81, 10))
plt.legend((p1[0], p2[0]), ('boys', 'girls'))

plt.show()
```
