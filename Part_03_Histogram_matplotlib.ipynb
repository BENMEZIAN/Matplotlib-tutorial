# What's a histogram
#### A Histogram represents the distribution of a numeric variable for one or several groups. The values are split in bins, each bin is represented as a bar.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data
data = np.random.randn(1000)
#print(data)

# Create the histogram
plt.hist(data, bins=30, color='blue', density=True)

# Add a KDE (Kernel Density Estimate) line on top of the histogram
sns.kdeplot(data, color='red', linewidth=2)

# Add labels and title
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Data')

# Show the plot
plt.show()
```

# multiple histograms at one figure

```python
import matplotlib.pyplot as plt
import numpy as np
import random
from matplotlib import style


ml_students_age = np.random.randint(18,45, (10))
py_students_age = np.random.randint(15,40, (10))
bins = [15,20,25,30,35,40,45] # category of ML students age on x axis

style.use("ggplot") # return grid
plt.figure(figsize = (16,9))

plt.hist([ml_students_age, py_students_age], bins, rwidth=0.8, histtype = "bar",orientation='vertical', color = ["m", "y"], label = ["ML Student", "Py Student"])

plt.title("ML & Py Students age histograms")
plt.xlabel("Students age cotegory")
plt.ylabel("No. Students age")
plt.legend()
plt.show()
```

# Hsitogram vs bar plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Generate some random data
data = np.random.randn(1000)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot a histogram
ax.hist(data, bins=30, density=True, alpha=0.6, color='blue')

# Add a KDE plot
sns.kdeplot(data, ax=ax, color='red', linewidth=2)

# Add labels and title
ax.set_xlabel('Value')
ax.set_ylabel('Density')
ax.set_title('Histogram with KDE')

# Show the plot
plt.show()
```

```python
#####################Differences:##########################
# Type of Data:
# - Histogram is used for numerical data (e.g., ages, incomes, test scores).
# - Bar plot is used for categorical data (e.g., fruits, countries, cities).

# X-Axis:
# - Histogram: X-axis represents numerical ranges (bins).
# - Bar plot: X-axis represents discrete categories (e.g., "Apples", "Bananas").

# Bars:
# - In a histogram, the bars represent frequency of data points in intervals.
# - In a bar plot, the bars represent frequency of categories.

#In Summary:
#A histogram is used to represent numerical data grouped into intervals, while a bar plot represents categorical data and shows the frequency of
#each category.
```
